```python
import numpy as np
import msprime
import tskit
import tscompare
```

## Simulate a tree sequence.

If you don't want anything special you can just use `msprime.sim_ancestry()`.

```python
ts = msprime.sim_ancestry(5, population_size=50, recombination_rate=1e-7, sequence_length=1e5)
```

```python
ts.draw_svg()
```

```python
# Extend Haplotypes
ets = ts.extend_haplotypes()
```

```python
ets.draw_svg()
```

### Compare

You can use `tscompare.haplotype_arf(ts, ets)` to compare the two ARGs and compute some statistics about them.

See [here](https://tskit.dev/tscompare/docs/stable/api.html#tscompare.haplotype_arf) for a detailed description of the output.

```python
# We can compare the two ARGs by using tscompare.compare (se)
tscompare.haplotype_arf(ts, ets)
```
